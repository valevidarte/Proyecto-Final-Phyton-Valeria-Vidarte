### Base de datos:

En el presente trabajo, se utilizará una base de datos sobre los Indicadores de Desarrollo Global a nivel país. El objetivo de este trabajo es realizar una comparación mediante gráficos entre los países de las regiones de Sudámerica, Centroamérica y Norteamérica a partir del siguiente indicador: Acceso a la electricidad rural (según el % de población rural de cada país) entre los años 2010 y 2021.

### 1. Importar la base de datos:

```python
# Importar las librerías pandas, altair y numpy:
import pandas as pd
import altair as alt
import numpy as np
```

```python
# Leer el archivo cvs del la base de datos:
locacion_datos = "51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1.csv"
IDG = pd.read_csv(locacion_datos, encoding='latin-1')
IDG.columns= IDG.columns.str.lower()
IDG.head()
```

### 2. Limpieza de los datos:

```python
# Reemplazamos los nombres de los códigos de los indicadores:
IDG['series code'] = IDG['series code'].replace({'EG.ELC.ACCS.ZS': 'electricidad','EG.ELC.ACCS.RU.ZS': 'electricidad_rural','EN.ATM.CO2E.PC': 'co2','NY.GDP.MKTP.KD.ZG': 'pbi','SH.MED.BEDS.ZS': 'camas_hospital','SL.TLF.TOTL.FE.ZS': 'fuerza_laboral_femenina'})

# Eliminamos las columnas antes de 2010:
IDG = IDG.drop(IDG.columns[4:54], axis=1)

# Nombra las columnas según su año en números:
IDG.columns.values[4:] = [str(i) for i in range(2010, 2023)] 

IDG.head(6)
```

### 3. Creación de subdatas para cada región:

#### Subdata: Acceso a la electricidad rural en Norteamérica

```python
elec_rural_norteamerica = IDG.loc[(IDG['country name'].isin(['Canada', 'Mexico', 'United States'])) & (IDG['series code'] == 'electricidad_rural')]
elec_rural_norteamerica = elec_rural_norteamerica.transpose().reset_index()
elec_rural_norteamerica.rename(columns={'index': 'Year'}, inplace=True)

columnas_norteamerica = ['Year', 'Canada', 'Mexico', 'United States']
elec_rural_norteamerica.columns = columnas_norteamerica
elec_rural_norteamerica = elec_rural_norteamerica.iloc[4:]
elec_rural_norteamerica
```

#### Subdata: Acceso a la electricidad rural en Centroamérica

```python
elec_rural_centroamerica = IDG.loc[(IDG['country name'].isin(['Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama'])) & (IDG['series code'] == 'electricidad_rural')]
elec_rural_centroamerica = elec_rural_centroamerica.transpose().reset_index()
elec_rural_centroamerica.rename(columns={'index':'Year'}, inplace=True)

columnas = ['Year','Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama']

elec_rural_centroamerica.columns = columnas
elec_rural_centroamerica = elec_rural_centroamerica.iloc[4:]
elec_rural_centroamerica
```

#### Subdata: Acceso a la electricidad rural en Sudamérica

```python
elec_rural_sudamerica = IDG.loc[(IDG['country name'].isin(['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela'])) & (IDG['series code'] == 'electricidad_rural')]
elec_rural_sudamerica = elec_rural_sudamerica.transpose().reset_index()
elec_rural_sudamerica.rename(columns={'index': 'Year'}, inplace=True)

columnas_sudamerica = ['Year', 'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela']

elec_rural_sudamerica.columns = columnas_sudamerica
elec_rural_sudamerica = elec_rural_sudamerica.iloc[4:]
elec_rural_sudamerica
```

### 3. Gráficos de las subdatas:

#### Gráfico: Acceso a la electricidad rural en Norteamérica

```python
# Gráfico de Canada:
graph_canada = alt.Chart(elec_rural_norteamerica).mark_line().encode(
    x=alt.X("Year:N", title="Year"),
    y=alt.Y("Canada:Q", title=""),
    color=alt.value("red"),
).interactive()


# Gráfico de México:
graph_mexico = alt.Chart(elec_rural_norteamerica).mark_line().encode(
    x=alt.X("Year:N", title="Year"),
    y=alt.Y("Mexico:Q", title=""),
    color=alt.value("green"),
).interactive()

# Gráfico de Unite States:
graph_usa = alt.Chart(elec_rural_norteamerica).mark_line().encode(
    x=alt.X("Year:N", title="Year"),
    y=alt.Y("USA:Q", title=""),
    color=alt.value("blue"),
).interactive()
```

```python
# Creación de la leyenda:
legend = alt.Chart(pd.DataFrame({'country': ['Canada', 'Mexico', 'United States'], 'color': ['red', 'green','blue']})).mark_point().encode(
    y=alt.Y('country:N', title=None),
    color=alt.Color('color:N', scale=None),
).properties(
    width=20)
```

```python
# Se junta la leyenda y gráficos:
elec_graph_norte = alt.concat(graph_canada+graph_mexico+graph_usa, legend, spacing=5, title="Acceso a la electricidad rural en Norteamérica")
```

```python
# Presentación del gráfico:
elec_graph_norte.display()   
```
